```python
import json
import mysql.connector
import pandas as pd

def connect_to_db():
    return mysql.connector.connect(
       host="localhost",
       user="root",
       password="",
       database="categories_lead"
    )


def fetch_leave_data(lead_id):
    try:
        conn = connect_to_db()
        cursor = conn.cursor(dictionary=True)
        query = """
        SELECT l.lead_id, l.inactivity_days, l.negative_sentiment_count, l.total_interactions
        FROM lead_engagement_summary as l
        WHERE l.lead_id = %s
        """
        cursor.execute(query, (lead_id,))
        result = cursor.fetchone()
        conn.close()
        return result if result else None
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        return None


def calculate_churn_score(lead_data):
    score = 0

    # 1. Negative Sentiment Score (Max 30)
    nsc = lead_data['negative_sentiment_count']
    if nsc == 1:
        score += 10
    elif nsc == 2:
        score += 20
    elif nsc >= 3:
        score += 30

    # 2. Inactivity Days (from 4th week onward)
    inactivity_dict = json.loads(lead_data['inactivity_days'])
    inactivity_from_week4 = sum([inactivity_dict.get(f"week {i}", 0) for i in range(4, 100)])
    if inactivity_from_week4 > 15:
        score += 35
    else:
        score += (inactivity_from_week4 / 15) * 35  # Scaled

    # 3. Interaction Score (from 4th week onward with recency-weighted penalty)
    interactions_dict = json.loads(lead_data['total_interactions'])

    bad_weight = 0
    total_weight = 0

    for week_key in interactions_dict:
        if week_key.startswith("week"):
            week_num = int(week_key.split()[1])
            if week_num >= 4:
                weight = week_num - 3  # Week 4 = 1, Week 5 = 2, etc.
                total_weight += weight
                if interactions_dict[week_key] <= 2:
                    bad_weight += weight

    if total_weight > 0:
        interaction_penalty = (bad_weight / total_weight) * 35
    else:
        interaction_penalty = 0

    score += interaction_penalty

    # Cap the score to 100
    score = min(score, 100)

    # Churn Status
    if score > 70:
        risk = "High Churn Risk"
    elif score >= 40:
        risk = "Medium Churn Risk"
    else:
        risk = "Low Churn Risk"

    return round(score, 2), risk


# Example usage
lead_id = 105
data = fetch_leave_data(lead_id)
if data:
    churn_score, status = calculate_churn_score(data)
    print(f"Lead {lead_id} → Churn Score: {churn_score}, Status: {status}")
else:
    print("Lead not found.")
```

```python
import mysql.connector
from datetime import datetime
import json

# Connect to DB
def connect_db():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="categories_lead"
    )

# Main logic
def calculate_weekly_calls():
    db = connect_db()
    cursor = db.cursor(dictionary=True)

    # Get all leads and their creation date
    cursor.execute("SELECT lead_id, created_at FROM leads")
    leads = cursor.fetchall()

    for lead in leads:
        lead_id = lead['lead_id']
        start_date = lead['created_at']

        # Get all call timestamps for this lead
        cursor.execute("""
            SELECT created_at FROM calls 
            WHERE lead_id = %s 
            ORDER BY created_at ASC
        """, (lead_id,))
        call_entries = cursor.fetchall()

        weekly_counts = {}
        for entry in call_entries:
            call_time = entry['created_at']
            delta = (call_time - start_date).days
            week_num = (delta // 7) + 1
            key = f"week {week_num}"
            weekly_counts[key] = weekly_counts.get(key, 0) + 1

        # Fill missing weeks with 0
        if weekly_counts:
            max_week = max(int(k.split()[1]) for k in weekly_counts)
            for i in range(1, max_week + 1):
                key = f"week {i}"
                if key not in weekly_counts:
                    weekly_counts[key] = 0

        # Ensure sorting by week number
        weekly_counts = dict(sorted(weekly_counts.items(), key=lambda x: int(x[0].split()[1])))

        weekly_json = json.dumps(weekly_counts)

        # Insert/update into interaction_count_calls
        cursor.execute("SELECT COUNT(*) as count FROM interaction_count_calls WHERE lead_id = %s", (lead_id,))
        exists = cursor.fetchone()['count']

        if exists:
            cursor.execute("""
                UPDATE interaction_count_calls 
                SET calls = %s 
                WHERE lead_id = %s
            """, (weekly_json, lead_id))
        else:
            cursor.execute("""
                INSERT INTO interaction_count_calls (lead_id, calls)
                VALUES (%s, %s)
            """, (lead_id, weekly_json))

        print(f"Lead {lead_id} → {weekly_json}")

    db.commit()
    db.close()

# Run
calculate_weekly_calls()
```

```python
import mysql.connector
from datetime import datetime
import json

# Connect to DB
def connect_db():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="categories_lead"
    )

# Main logic
def calculate_weekly_meetings_from_emails():
    db = connect_db()
    cursor = db.cursor(dictionary=True)

    # Get all leads and their creation date
    cursor.execute("SELECT lead_id, created_at FROM leads")
    leads = cursor.fetchall()

    for lead in leads:
        lead_id = lead['lead_id']
        start_date = lead['created_at']
        if isinstance(start_date, str):
            start_date = datetime.strptime(start_date, '%Y-%m-%d %H:%M:%S')

        # Get all "received" email timestamps for this lead
        cursor.execute("""
            SELECT timestamp FROM emails 
            WHERE lead_id = %s AND direction = 'received'
            ORDER BY timestamp ASC
        """, (lead_id,))
        email_entries = cursor.fetchall()

        weekly_counts = {}

        for entry in email_entries:
            email_time = entry['timestamp']
            if isinstance(email_time, str):
                email_time = datetime.strptime(email_time, '%Y-%m-%d %H:%M:%S')
            delta = (email_time - start_date).days
            week_num = max((delta // 7) + 1, 1)  # Ensure minimum is week 1
            key = f"week {week_num}"
            weekly_counts[key] = weekly_counts.get(key, 0) + 1

        # Fill missing weeks
        if weekly_counts:
            max_week = max(int(k.split()[1]) for k in weekly_counts)
            for i in range(1, max_week + 1):
                key = f"week {i}"
                if key not in weekly_counts:
                    weekly_counts[key] = 0

        # Ensure proper order
        weekly_counts = dict(sorted(weekly_counts.items(), key=lambda x: int(x[0].split()[1])))
        weekly_json = json.dumps(weekly_counts)

        # Check if lead already exists in interaction_counts
        cursor.execute("SELECT COUNT(*) as count FROM interaction_counts WHERE lead_id = %s", (lead_id,))
        exists = cursor.fetchone()['count']

        if exists:
            cursor.execute("""
                UPDATE interaction_counts
                SET meetings = %s
                WHERE lead_id = %s
            """, (weekly_json, lead_id))
        else:
            cursor.execute("""
                INSERT INTO interaction_counts (lead_id, calls, meetings)
                VALUES (%s, %s, %s)
            """, (lead_id, '{}', weekly_json))

        print(f"Lead {lead_id} → {weekly_json}")
    db.commit()
    db.close()
calculate_weekly_meetings_from_emails()
```

```python
#inactivity  days 

import mysql.connector
from datetime import datetime, timedelta

def connect_db():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="categories_lead"
    )

def update_inactivity_days():
    db = connect_db()
    cursor = db.cursor(dictionary=True)

    # Step 1: Determine current Sunday
    today = datetime.today()
    current_sunday = today + timedelta(days=(6 - today.weekday())) if today.weekday() < 6 else today
    current_sunday = current_sunday.replace(hour=0, minute=0, second=0, microsecond=0)

    # Step 2: Get all leads
    cursor.execute("SELECT lead_id, created_at FROM leads")
    leads = cursor.fetchall()

    for lead in leads:
        lead_id = lead["lead_id"]
        created_at = lead["created_at"]
        if isinstance(created_at, str):
            created_at = datetime.strptime(created_at, "%Y-%m-%d %H:%M:%S")

        # Step 3: Get most recent interaction (emails, calls, meetings)
        cursor.execute("""
            SELECT MAX(ts) as latest FROM (
                SELECT MAX(timestamp) as ts FROM emails 
                    WHERE lead_id = %s AND direction = 'received'
                UNION
                SELECT MAX(created_at) as ts FROM calls 
                    WHERE lead_id = %s
                UNION
                SELECT MAX(created_at) as ts FROM meetings 
                    WHERE related_to = %s
            ) as combined
        """, (lead_id, lead_id, lead_id))

        latest = cursor.fetchone()["latest"]

        if latest:
            if isinstance(latest, str):
                latest = datetime.strptime(latest, "%Y-%m-%d %H:%M:%S")
            inactivity_days = (current_sunday.date() - latest.date()).days
        else:
            inactivity_days = 7  # assume full week inactive

        # Step 4: Insert or update inactivity_days table
        cursor.execute("SELECT COUNT(*) as count FROM inactivity_days WHERE lead_id = %s", (lead_id,))
        exists = cursor.fetchone()["count"]

        if exists:
            cursor.execute("""
                UPDATE inactivity_days
                SET inactivity_days = %s
                WHERE lead_id = %s
            """, (inactivity_days, lead_id))
        else:
            cursor.execute("""
                INSERT INTO inactivity_days (lead_id, inactivity_days)
                VALUES (%s, %s)
            """, (lead_id, inactivity_days))

        print(f"✅ Lead {lead_id}: {inactivity_days} days inactive")

    db.commit()
    db.close()

# Execute the script
update_inactivity_days()
```
